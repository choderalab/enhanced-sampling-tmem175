## Use plotly instead

```python
import plotly.express as px
import os
import pandas as pd
import numpy as np
```

```python
arrays = []
cols = []
bias_file_dir = "/Volumes/Rohirrim/local_test/enhanced_sampling/metadynamics_rmsd/sim01/bias-files/"
for file_name in os.listdir(bias_file_dir):
    file_path = os.path.join(bias_file_dir, file_name)
    cols.append(os.path.splitext(file_name)[0].split("_")[-1])
    vals = np.load(file_path)
    arrays.append(vals)
```

```python
cols
```

```python
df = pd.DataFrame(arrays).T
# df["rmsd"] = np.linspace(0.1, 4, 98)
```

```python
df
```

```python
df.index = np.linspace(0.1, 4, 98)
```

```python
df.columns = cols
```

```python
tidydf = df.stack()
tidydf
```

```python
tidydf.index = tidydf.index.rename('rmsd', level=0)
tidydf.index = tidydf.index.rename('time', level=1)
tidydf.name = 'bias'
```

```python
tdf = tidydf.reset_index()
```

```python
tdf
```

```python
px.line(tdf, 
        x='rmsd', 
        y='bias', 
        animation_frame='time')
```

```python
px.line(tdf, 
        x='rmsd', 
        y='bias', 
        color='time',
        category_orders={'time': cols}
       )
```
